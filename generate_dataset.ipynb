```python
# Import libraries
import pandas as pd
import pandas_profiling as pp
import matplotlib.pyplot as plt
import numpy as np
import sys
import matplotlib
import math
%matplotlib inline
```

```python
def read_portfolio(portfolio_num):

    df= pd.read_csv('portfolios\portfolio_' + str(portfolio_num) + '.csv')
        

    #create dictionary assingnment
    
    value_ID = portfolio_num
    value_VATD_NA = (df['Vehicle_Anti_Theft_Device'] == 'Not Applicable').sum()/df.shape[0]
    value_VATD_PASSIVE = (df['Vehicle_Anti_Theft_Device'] == 'Passive Disabling-Vehicle Recovery').sum()/df.shape[0]
    value_VATD_ALARM = (df['Vehicle_Anti_Theft_Device'] == 'Alarm Only').sum()/df.shape[0]
    value_VATD_ACTIVE = (df['Vehicle_Anti_Theft_Device'] == 'Active Disabling').sum()/df.shape[0]
    value_DMA_mean = (df['Driver_Minimum_Age']).mean()
    value_DMA_std = (df['Driver_Minimum_Age']).std()
    value_VAY_mean = (df['Vehicle_Age_In_Years']).mean()
    value_VAY_std = (df['Vehicle_Age_In_Years']).std()
    value_AP_mean = (df['Annual_Premium']).mean()
    value_AP_std = (df['Annual_Premium']).std()
    value_ln_LR = (math.log((df['Loss_Amount']).sum()/df.shape[0]/((df['Annual_Premium']).sum()/df.shape[0]),2))
    dictionary = {
        'ID': value_ID,
        'VATD_NA':value_VATD_NA,
        'VATD_PASSIVE': value_VATD_PASSIVE,
        'VATD_ALARM': value_VATD_ALARM,
        'DMA_mean': value_DMA_mean,
        'DMA_std': value_DMA_std,
        'VAY_mean': value_VAY_mean,
        'VAY_std': value_VAY_std,
        'AP_mean': value_AP_mean,
        'AP_std': value_AP_std,
        'ln_LR': value_ln_LR,
    }
    
    return dictionary
print(read_portfolio(1))


    

    







```

```python
data = []
for i in range(1,501):
    data.append(dict(read_portfolio(i)))
df = pd.DataFrame(data).set_index('ID')
        
df.head()
```
